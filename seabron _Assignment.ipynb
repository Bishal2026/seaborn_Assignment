***Q1***

Scatter Plot:
    Visualize the relationship between two continuous variables.
    
Line Plot:
    Show the trend or pattern of one or more variables over time.
    
Histogram:
    Display the distribution of a single variable.
    
Bar Plot:
    Compare categorical variables and their associated numerical values.
    
Heatmap:
    Visualize the correlation or relationship between multiple variables.
    

***Q2***

```python
import seaborn as sns
fmri = sns.load_dataset('fmri')
sns.lineplot( x="timepoint", y="signal",data=fmri, hue="event", style="region")
```

```python
import seaborn as sns
diamonds_data = sns.load_dataset("diamonds")
sns.histplot(data=diamonds_data, x='price', hue='cut', multiple="stack")
```

```python
import seaborn as sns
flights_data = sns.load_dataset("flights")
flights_matrix = flights_data.pivot(index="month", columns="year", values="passengers")

sns.heatmap(data=flights_matrix)
```
